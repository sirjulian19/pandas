```python
import pandas as pd
import numpy as np
d={'id':pd.Series([1,2,3,4,5]),
    'A':pd.Series(['A1','A2','A3','A4','A5'])
   }
df = pd.DataFrame(d)

d1={'id':pd.Series([1,2,7,11,15]),
    'K':pd.Series(['K1','K2','K3','K4','K5'])
   }
df1 = pd.DataFrame(d1)
```

```python
pd.merge(df,df1,on='id')
```

```python
pd.merge(df,df1,left_on='id',right_on='id1')
```

```python
pd.merge(df,df1,on='id',how='left')
```

```python
pd.merge(df,df1,on='id',how='right')
```

```python
df_1 = pd.merge(df, df1, on='id', how="outer", indicator=True
              ).query('_merge=="left_only"')
df_1
```

```python
df__1 = pd.merge(df, df1, on='id', how="outer", indicator=True
              ).query('_merge=="right_only"')
df__1
```
